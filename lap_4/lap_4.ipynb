```python
import cv2
# Load the image in grayscale
img = cv2.imread('grass.png', cv2.IMREAD_GRAYSCALE)
# Set the threshold values
threshold_values = [0, 50, 100, 150, 200]
# Apply thresholding with different threshold values
for threshold_value in threshold_values:
    ret, thresh = cv2.threshold(img, threshold_value, 255,
    cv2.THRESH_BINARY)
    cv2.imshow(f'Thresholded Image ({threshold_value})',
    thresh)
# Display the original image
cv2.imshow('Original Image', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Step 1: Load necessary libraries
import matplotlib.pyplot as plt
import numpy as np
from skimage import data
from skimage import exposure

# Step 2: Load moon image
img = data.moon()

# Step 3: Rescale intensity values between 2nd and 98th percentiles
# Contrast stretching
p2, p98 = np.percentile(img, (2, 98))
img_rescale = exposure.rescale_intensity(img, in_range=(p2, p98))

# Step 4: Display original image and its histogram
plt.figure(figsize=(10, 7))

plt.subplot(2, 1, 1)
plt.imshow(img, cmap='gray')
plt.axis('off')
plt.title('Original Image')

plt.subplot(2, 1, 2)
plt.hist(img.flatten(), bins=256, range=(0, 255))
plt.xlabel('Gray Level Values')
plt.ylabel('Number of Pixels')

plt.tight_layout()
plt.show()

# Display contrast-stretched image and its histogram
plt.figure(figsize=(10, 7))

plt.subplot(2, 1, 1)
plt.imshow(img_rescale, cmap='gray')
plt.axis('off')
plt.title('Contrast Stretched Image')

plt.subplot(2, 1, 2)
plt.hist(img_rescale.flatten(), bins=256, range=(0, 255))
plt.xlabel('Gray Level Values')
plt.ylabel('Number of Pixels')

plt.tight_layout()
plt.show()
```

### task 1

```python
from skimage import data
# Load the moon image (grayscale image)
img = data.moon()

# Compute the 3rd and 80th percentile intensity values
# These values will define the new intensity range
p3, p80 = np.percentile(img, (3, 80))

# Apply contrast stretching
# Intensities below p3 become darker
# Intensities above p80 become brighter
img_rescale = exposure.rescale_intensity(img, in_range=(p3, p80))

# Display the rescaled image and its histogram
plt.figure(figsize=(10, 7))

# Show processed image
plt.subplot(2, 1, 1)
plt.imshow(img_rescale, cmap='gray')
plt.axis('off')
plt.title('Rescaled Image (3rd - 80th Percentile)')

# Show histogram of processed image
plt.subplot(2, 1, 2)
plt.hist(img_rescale.flatten(), bins=256, range=(0, 255))
plt.xlabel('Gray Level Values')
plt.ylabel('Number of Pixels')

plt.tight_layout()
plt.show()
```

```python

# Load the moon image
img = data.moon()

# Apply histogram equalization
# This redistributes pixel intensities to enhance contrast
img_equalized = exposure.equalize_hist(img)

# Display equalized image and its histogram
plt.figure(figsize=(10, 7))

# Show equalized image
plt.subplot(2, 1, 1)
plt.imshow(img_equalized, cmap='gray')
plt.axis('off')
plt.title('Histogram Equalized Image')

# Show histogram after equalization
# Note: values are between 0 and 1 (float image)
plt.subplot(2, 1, 2)
plt.hist(img_equalized.flatten(), bins=256)
plt.xlabel('Gray Level Values')
plt.ylabel('Number of Pixels')

plt.tight_layout()
plt.show()
```

### Task3

```python
from skimage.exposure import match_histograms

# Load source image (Chelsea - cat image)
# This image will be transformed
source = data.chelsea()

# Load reference image (Rocket image)
# The source image will match this histogram
reference = data.rocket()

# Apply histogram matching
# channel_axis=-1 ensures matching is done per color channel (RGB)
matched = match_histograms(source, reference, channel_axis=-1)

# Display source, reference, and matched images
plt.figure(figsize=(12, 8))

# Original source image
plt.subplot(3, 1, 1)
plt.imshow(source)
plt.axis('off')
plt.title('Source Image (Chelsea)')

# Reference image
plt.subplot(3, 1, 2)
plt.imshow(reference)
plt.axis('off')
plt.title('Reference Image (Rocket)')

# Result after histogram matching
plt.subplot(3, 1, 3)
plt.imshow(matched)
plt.axis('off')
plt.title('Histogram Matched Image')

plt.tight_layout()
plt.show()
```
